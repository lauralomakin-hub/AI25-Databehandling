```python
import pandas as pd
import matplotlib .pyplot as plt

athletes = pd.read_csv("../data/athlete_events.csv")
region = pd.read_csv("../data/noc_regions.csv")

region.head()

athletes.head()

merged = pd.merge(athletes, region, on="NOC", how="outer")   #"outer" för att inte tappa någon data från atbellerna. saknade värden fylls med NaN

merged.tail()
```

```python
#skapa tabeller för NOC. KOntrollera om det finns unika värden ej kopplade till länder.
# Vi använder .unique() för att hämta alla unika NOC-koder från kolumnen 
# (utan dubbletter, i samma ordning som de förekommer i datan).
# Sedan gör vi om resultatet till ett set() för att kunna jämföra mängder
# mellan tabeller (t.ex. hitta NOC-koder som saknas i regions-filen).
# Kombinationen ger en effektiv och tydlig jämförelse av unika värden.

athlete_nocs = set(athletes["NOC"].unique())
region_nocs = set(region["NOC"].unique())

missing_nocs = athlete_nocs - region_nocs

missing_nocs
```

Our NOC code is now "SGP" instead of "SIN". In line with the United Nations (UN) / International Organisation for Standardisation (ISO) official 3-letter country code, the International Olympic Committee (IOC) has approved SNOC’s application to change our NOC code from “SIN” to “SGP”. The first major Games which the code will take effect would be at the Sapporo 2017 Asian Winter Games.
