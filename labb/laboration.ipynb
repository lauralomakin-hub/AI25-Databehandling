Jag kan därmend inte göra en analys jag skulle gjort för ett svenskt dataset men väljer att undersöka Grad.rate mot "PhD : Percent of faculty with Ph.D.s" (som jag tolkar som andel lärare med forskarexamen) och "Terminal : Percent of faculty with terminal degree" (som jag tolkar som högsta examen  i sitt ämne.)

```python
#Frågar Mistral AI hur jag kan använda trendline="ols" i dessa subplots:
#"Du kan använda numpy för att beräkna regressionslinjen och sedan rita den med matplotlib.

def add_ols_trendline(x,y, ax):  # En funktion som tar parametrarna x, y och ax (en matplotlib-axel där trendline ska ritas)
    z = np.polyfit(x, y, 1)      # Beräknar koefficienterna för en linjär regressionslinje
    p=np.poly1d(z)               #p (linjen för y=kx+m) =  poly1d (Tar koefficienten z, skapar en funktion som kan beräkna y-värden för givna x-värden)
    ax.plot(x, p(x), color="red")
fig, axes = plt.subplots(1, 2, figsize=(12, 5))   

axes[0].scatter(college["PhD"], college["Grad.Rate"])
axes[0].set_title("Faculty member with a PhD vs. Graduation rate")
axes[0].set_xlabel("PhD")
axes[0].set_ylabel("Graduation rate")
add_ols_trendline(college["PhD"], college["Grad.Rate"], axes[0])       #anropar add_ols_trendlines med tre argument

axes[1].scatter(college["Terminal"], college["Grad.Rate"])
axes[1].set_title("Faculty member with a terminal degre vs. Graduation rate")
axes[1].set_xlabel("Terminal degree")
axes[1].set_ylabel("Graduation rate")
add_ols_trendline(college["Terminal"], college["Grad.Rate"], axes[1]) 
```
