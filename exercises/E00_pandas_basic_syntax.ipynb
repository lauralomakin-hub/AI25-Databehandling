````python
# 1) Importer
import pandas as pd
import matplotlib.pyplot as plt

# 2) (Som om från API) – exempeldata i samma form som ett JSON-svar
data = [
    {"date": "2025-10-30", "open": 100.0, "close": 103.5},
    {"date": "2025-10-31", "open": 103.5, "close": 101.0},
    {"date": "2025-11-01", "open": 101.0, "close": 105.8},
    {"date": "2025-11-02", "open": 105.8, "close": 104.6},
    {"date": "2025-11-03", "open": 104.6, "close": 108.2},
]

# 3) In i en DataFrame
df = pd.DataFrame(data)

# 4) Gör om text → datum så vi kan sortera/filtrera korrekt
df["date"] = pd.to_datetime(df["date"])

# (valfritt men bra för tidsserier)
df = df.sort_values("date").reset_index(drop=True)

# 5) Beräkna daglig förändring (absolut och procent)
df["daily_change"] = df["close"] - df["open"]
df["daily_change_pct"] = (df["close"] / df["open"] - 1) * 100

# 6) Snabb statistik
min_change = df["daily_change"].min()
max_change = df["daily_change"].max()
mean_change = df["daily_change"].mean()

print("== Daglig förändring (absolut) ==")
print("Min:", round(min_change, 2))
print("Max:", round(max_change, 2))
print("Mean:", round(mean_change, 2))

# 7) Plotta daglig förändring över tid
ax = df.plot(x="date", y="daily_change", title="Daily Stock Changes")

# Lägg in hjälplinjer för min/medel/max (påverkar grafen visuellt)
import matplotlib.pyplot as plt
plt.axhline(mean_change, linestyle="--", label="Mean")
plt.axhline(max_change, linestyle="--", label="Max")
plt.axhline(min_change, linestyle="--", label="Min")

plt.legend()
plt.xlabel("Date")
plt.ylabel("Daily change (close - open)")
plt.show()

# 8) (valfritt) – titta även procentuellt
print("\n== Daglig förändring (%) ==")
print(df[["date", "daily_change_pct"]])
````
